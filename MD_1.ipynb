```python
import pandas as pd

# Create a dictionary to store all dataframes
all_dataframes = {}

# List all files in the data directory
files = pd.read_csv("Term-paper/data")  # This will show all files in directory
csv_files = [f for f in files if f.endswith('.csv')]

# Read each CSV file into a dataframe
for file in csv_files:
    # Get filename without extension
    filename = file[:-4]  # Remove '.csv'
    # Read the CSV file
    all_dataframes[filename] = pd.read_csv(f"Term-paper/data/{file}")

# Print the names of loaded files and their shapes
for name, df in all_dataframes.items():
    print(f"Loaded {name}: {df.shape} rows Ã— columns")
```
